```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset into a pandas DataFrame
df = pd.read_csv('Data.csv')
df
```

```python


# Create a user-tweet matrix
user_tweet_matrix = df.pivot_table(index='user_id', columns='tweet_id', values='likes', fill_value=0)

# Calculate cosine similarity between tweets
tweet_similarity = cosine_similarity(user_tweet_matrix.T)

# Create a function to get top recommended tweets
def get_top_tweets(likes, n=5):
    tweet_index = np.where(user_tweet_matrix.columns == likes)[0][0]
    similar_tweets = tweet_similarity[tweet_index]
    similar_tweet_indices = np.argsort(-similar_tweets)[:n]
    return list(user_tweet_matrix.columns[similar_tweet_indices])

# Example usage
top_tweets = get_top_tweets('1432393529610908677', n=3)
print(top_tweets)
```

```python
import pandas as pd

# Read data from CSV
df = pd.read_csv("Data.csv")

# Group by hashtag and location, get total likes and sort by descending order
hashtag_likes = df.groupby("hashtags")["likes"].sum().reset_index().sort_values(by=["likes"], ascending=False)
location_likes = df.groupby("user_location")["likes"].sum().reset_index().sort_values(by=["likes"], ascending=False)

# Print top 5 hashtag and location recommendations by likes
print("Hashtag recommendations by likes:")
print(hashtag_likes.head())
print("\nLocation recommendations by likes:")
print(location_likes.head())

# Combine hashtag and location recommendations
combined_likes = pd.concat([hashtag_likes, location_likes]).groupby(["hashtags", "user_location"])["likes"].sum().reset_index().sort_values(by=["likes"], ascending=False)

# Print top 5 combined recommendations by likes
print("\nCombined recommendations by likes:")
print(combined_likes.head())
```
